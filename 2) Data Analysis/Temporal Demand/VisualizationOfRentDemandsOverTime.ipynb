## Visualization Of Temporal Demand Patterns and Seasonality ##

The following script is a continuation of previous script 'Temporal Demand Patterns and Seasonality'. It uses the 4 generated CSV datasets and visualizes them in one picture. Optional feature is to save the resulted graphic as a PNG and PDF.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Read the 4 CSV documents generated previously
RentsPerDaytime = pd.read_csv("RentsPerDaytime_2019.csv")
RentsPerWeekday = pd.read_csv("RentsPerWeekday_2019.csv")
RentsPerMonth = pd.read_csv("RentsPerMonth_2019.csv")
RentsPerCalendarWeek = pd.read_csv("RentsPerCalendarWeek_2019.csv")

# Generate structure of whole image
fig = plt.figure(figsize=(25, 15))
fig.suptitle("Temporal Demand Patterns and Seasonality", fontsize=28, fontweight='bold')
#plt.subplots_adjust(wspace= 0.25, hspace= 0.25)

# Design of first subplot, Rents Per Daytime:
subplotRentsPerDaytime = fig.add_subplot(2,3,1)
subplotRentsPerDaytime.set_title('Rents Per Daytime', fontsize=20)
subplotRentsPerDaytime.set_ylabel('Rents', rotation = 0, position=(0,1.02), fontsize=14)
subplotRentsPerDaytime.set_xlabel('Daytime in h', rotation = 0, fontsize=14)
subplotRentsPerDaytime.set_xlim([-1,24])
ax_RentsPerDaytime = np.arange(0, 24)
subplotRentsPerDaytime.set_ylim([min(RentsPerDaytime['Count'])-20000, max(RentsPerDaytime['Count'])+20000])
subplotRentsPerDaytime.set_xticks(np.arange(min(ax_RentsPerDaytime), max(ax_RentsPerDaytime)+1, 1.0))
subplotRentsPerDaytime.bar(ax_RentsPerDaytime,RentsPerDaytime['Count'], color='orange')

# Design of second subplot, Rents Per Weekdays:
subplotRentsPerWeekday = fig.add_subplot(2,3,2)
subplotRentsPerWeekday.set_title('Rents Per Weekday', fontsize=20)
subplotRentsPerWeekday.set_ylabel('Rents', rotation = 0, position=(0,1.02), fontsize=14)
subplotRentsPerWeekday.set_xlabel('Weekday', rotation = 0, fontsize=14)
subplotRentsPerWeekday.set_xlim([-1,7])
ax_RentsPerWeekday = np.arange(0, 7)
subplotRentsPerWeekday.set_ylim([min(RentsPerWeekday['Count'])-20000, max(RentsPerWeekday['Count'])+15000])
subplotRentsPerWeekday.set_xticks(np.arange(min(ax_RentsPerWeekday), max(ax_RentsPerWeekday)+1, 1.0))
subplotRentsPerWeekday.set_xticklabels(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
subplotRentsPerWeekday.bar(ax_RentsPerWeekday,RentsPerWeekday['Count'], color='green')

# Design of third subplot, Rents Per Month:
subplotRentsPerMonth = fig.add_subplot(2,3,3)
subplotRentsPerMonth.set_title('Rents Per Month', fontsize=20)
subplotRentsPerMonth.set_xlabel('Month', rotation = 0, fontsize=14)
subplotRentsPerMonth.set_ylabel('Rents', rotation = 0, position=(0,1.02), fontsize=14)
subplotRentsPerMonth.set_xlim([-1,12])
ax_RentsPerMonth = np.arange(0, 12)
subplotRentsPerMonth.set_ylim([min(RentsPerMonth['Count'])-20000, max(RentsPerMonth['Count'])+20000])
subplotRentsPerMonth.set_xticks(np.arange(min(ax_RentsPerMonth), max(ax_RentsPerMonth)+1, 1.0))
subplotRentsPerMonth.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
subplotRentsPerMonth.bar(ax_RentsPerMonth,RentsPerMonth['Count'], color='darkred')

# Design of fourth subplot, Rents Per Calendar Week, 
# as the x-axis has 52 ticks it is a good idea to stretch the plot over three 'cells':
subplotRentsPerCalendarWeek = fig.add_subplot(2,3,(4,6))
subplotRentsPerCalendarWeek.set_title('Rents Per Calendar Week', fontsize=20)
subplotRentsPerCalendarWeek.set_xlabel('Week number', rotation = 0, fontsize=14)
subplotRentsPerCalendarWeek.set_ylabel('Rents', rotation = 0, position=(0,1.02), fontsize=14)
subplotRentsPerCalendarWeek.set_xlim([0,53])
ax_RentsPerCalendarWeek = np.arange(1, 53)
subplotRentsPerCalendarWeek.set_ylim([min(RentsPerCalendarWeek['Count'])-10000, max(RentsPerCalendarWeek['Count'])+10000])
subplotRentsPerCalendarWeek.set_xticks(np.arange(min(ax_RentsPerCalendarWeek), max(ax_RentsPerCalendarWeek)+1, 1.0))
subplotRentsPerCalendarWeek.bar(ax_RentsPerCalendarWeek,RentsPerCalendarWeek['Count'], color='lightblue')
print()
```

```python
# Save the picture as PNG and PDF file
fig.savefig('DemandPatterns.pdf', dpi=300)
fig.savefig('DemandPatterns.png', dpi=300)
```
